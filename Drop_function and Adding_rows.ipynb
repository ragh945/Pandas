# Drop function and Adding of rows at specified index

# drop():
The drop() method removes the specified row or column.

By specifying the column axis (axis='columns'), the drop() method removes the specified column.

By specifying the row axis (axis='index'), the drop() method removes the specified row.

# Syntax:
```dataframe.drop(labels, axis, index, columns, level, inplace., errors)```

# Dropping of columns

# In drop()--------axis=1(columns)  and axis=0(rows)

df.drop(
    labels=None,
    axis: 'Axis' = 0,
    index=None,
    columns=None,
    level: 'Level | None' = None,
    inplace: 'bool' = False,
    errors: 'str' = 'raise')

````python
df1.drop(columns="type")
````

````python
df1.drop(columns=["type","imdb_id"])
````

````python
df1.drop("type",axis=1) #both are same with columns,axis=1 gives the same answer
````

````python
df1.drop(["type","id"],axis=1)
````

````python
df1.drop(df1.loc[:,["type","id"]],axis=1)
````

````python
df1.drop(df1.loc[:,"id":"type"],axis=1)
````

````python
df1.drop(df1.iloc[:,0:3],axis=1)
````

# Adding of rows using concat()

# df1.sort_index().reset_index(drop=True)

````python
IDs=["INO001","INO002","INO003","INO004","INO005","INO006","INO007","INO008","INO009","INO010"]
Names=["Raghu","Vamshi","Irfan","Khan","John","Shubham","Vijith","Gurucharan","Shravan","Pavan"]
Age=[23,23,23,23,24,23,24,23,23,23]
Gender=["M","M","M","M","M","M","M","M","M","M"]
Marks=[15,85,70,96,88,66,100,75,97,60]
````

````python
d1=pd.DataFrame({"sid":IDs,"sname":Names,"age":Age,"gender":Gender,"marks":Marks})
d2=d1.copy()
d2
````

````python
dict1={"sid":"INO011","sname":"Mukesh","new_name":"MUKESH","age":24,"gender":"M","marks":92}
d3=pd.DataFrame(dict1,index=[0])
d3
````

````python
#Adding row at the last using concat()
d4=pd.concat([d2,d3],ignore_index=True)
d4
````

````python
data1=pd.DataFrame({"sid":["INO012","INO013"],"sname":["Vinay","Rohit"],"new_name":["VINAY","ROHIT"],"age":[24,23],"gender":["M","M"],"marks":[84,99]})
````

````python
data1
````

````python
#Adding 2 rows at the begining using concat()
d5=pd.concat([data1,d4],ignore_index=True)
d5
````

````python
#Adding row at 2nd index position
d5.loc[1.5]=["IN014","Virat","VIRAT",35,"M",91]
````
