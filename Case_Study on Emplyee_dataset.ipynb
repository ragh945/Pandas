# Filtering of columns based on different conditions

# What are the target_names whose sepallength<4.6 and petal length>1.5 

```python
d2.loc[(d2["sepal length (cm)"]>4.6) & (d2["petal length (cm)"]>1.5),["sepal length (cm)","petal length (cm)","target_names"]]
```

# Filter the dataframe  whose sepallength<4.9 and petal length>1.5 belongs to setosa

```python
d2.loc[(d2["sepal length (cm)"]<4.9) & (d2["petal length (cm)"]>1.5) & (d2["target_names"]=="setosa"),["sepal length (cm)","petal length (cm)","target_names"]]
```

# Filter the dataframe  whose sepallength<7.5 and petal length>6.2 belongs to virginica

```python
d2.loc[(d2["sepal length (cm)"]<7.5) & (d2["petal length (cm)"]>6.2) & (d2["target_names"]=="virginica"),["sepal length (cm)","petal length (cm)","target_names"]]
```

# Filter the dataframe  whose sepallength<7and petal length>5 belongs to versicolor

```python
d2.loc[(d2["sepal length (cm)"]<7) & (d2["petal length (cm)"]>5) & (d2["target_names"]=="versicolor"),["sepal length (cm)","petal length (cm)","target_names"]]
```
